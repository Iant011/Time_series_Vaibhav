## Time resampling

```python
df_Tsl
```

```python
df_Tsl.resample(rule='A')
```

```python
# DatetimeIndexResampler
```

```python
# A rule year end frequency
df_Tsl.resample(rule='A').min()
```

```python
df_Tsl.resample(rule='A').max()
```

```python
df_Tsl.resample(rule='A').max()['Open'].plot()
```

```python
## If we need quaterly
df_Tsl.resample(rule='QS').max()['Open'].plot()
```

```python
df_Tsl.resample(rule='QS').max()['High'].plot()
```

```python
df_Tsl.resample(rule='BA').max()['Open'].plot()
```

```python
df_Tsl.resample(rule='BA').max()
```

```python
df_Tsl.resample(rule='BQS').max()['Open'].plot()
```

```python
# plotting
df_Tsl['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_Tsl['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# monthly max data
df_Tsl['Open'].resample(rule='M').max()
```

```python
df_Tsl['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```
