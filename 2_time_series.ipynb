```python
import pandas_datareader as pdr
import pandas as pd
import datetime
from datetime import datetime
from pandas_datareader import data as pdr
import yfinance as yfin
```

```python
yfin.pdr_override()

df_Tsl = pdr.get_data_yahoo('TSLA', start='2019-10-24', end='2023-8-21')

df_Tsl
```

```python
# 1.Simple Moving Average
```

```python
df_Tsl.plot(figsize=(15,6))
```

```python
df_Tsl['Open:10Days']=df_Tsl['Open'].rolling(window=10).mean()
```

```python
df_Tsl[['Open','Open:10Days']].plot(figsize=(15,6))
```

```python
df_Tsl['Open:50Days']=df_Tsl['Open'].rolling(window=50).mean()
df_Tsl['Open:200Days']=df_Tsl['Open'].rolling(window=200).mean()
```

```python
df_Tsl[['Open','Open:50Days','Open:200Days']].plot(figsize=(15,6))
```

```python
## Expanding
## CMA
df_Tsl['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
# EWMA Epo Weighted Moving Avg.
# Focus on the new upcoming data
```

```python
# Ema on tesla
# Smoothning factor - 0.1
```

```python
df_Tsl['Ema_0.1']=df_Tsl['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df_Tsl[['Open','Ema_0.1']].plot(figsize=(15,6))
```

```python
df_Tsl['Ema_0.3']=df_Tsl['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
df_Tsl[['Open','Ema_0.1','Ema_0.3']].plot(figsize=(15,6))
```

```python
## Emplemetation of EWMA
df_Tsl['Ewma_05']=df_Tsl['Open'].ewm(span=9).mean()
```

```python
df_Tsl[['Open','Ewma_05']].plot(figsize=(15,6))
```
